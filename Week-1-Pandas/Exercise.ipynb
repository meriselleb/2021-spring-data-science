## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


```python
# 1. How many listings are there with a price less than 100? 
condition_less100 = df['price'] < 100
df[condition_less100]
df[condition_less100].shape[0] # counts rows
# 22,778 listings with a price less than 100
```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.
condition_bk = df['neighbourhood_group'] == 'Brooklyn'
df_bk = df[condition_bk]
total_listings_bk = df_bk.shape[0]
total_listings_bk
# 18,632 listings in brooklyn
```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100.

# method 1: using two methods on original df
method1result = df[condition_less100 & condition_bk].shape[0]
print(method1result, ' listings in Brooklyn with a price less than 100.')

# method 2: using df_bk 
condition_less100bk = df_bk['price'] < 100
method2result = df_bk[condition_less100bk].shape[0]
print(method2result, ' listings in Brooklyn with a price less than 100.')

# 10,473 listings in brooklyn with a price less than 100
```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
list_of_host_names = ['Michael', 'David', 'Daniel', 'John']
conditionNames = df.host_name.isin(list_of_host_names)
df[conditionNames]
```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 

condition_wb = df['neighbourhood']  == 'Williamsburg'
df['adjusted_price'] = np.where(condition_wb, df['price'] + 100, df['price'])

df.head()        # shows 'adjusted price' is the same for non-williamsburg listings
df[condition_wb] # to show 'adjusted price' for williamsburg listings since none were at the head
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.  

condition_private = df['room_type'] == 'Private room'
condition_shared = df['room_type'] == 'Shared room'

total_count = df.shape[0]
total_private = df[condition_private].shape[0]
total_shared = df[condition_shared].shape[0]
percent_private = (total_private / total_count) * 100
percent_shared = (total_shared / total_count) * 100

print('Percent Private: ', round(percent_private, 2), '%')
print('Percent Shared: ', round(percent_shared, 2), '%')
```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.

gbcount = df.groupby('neighbourhood_group').size()
gbcount.head()
grouped_df = pd.DataFrame(gbcount)
grouped_df = grouped_df.reset_index()
grouped_df.head()
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 
gbprice = df.groupby('neighbourhood_group')['price'].mean()
gbprice_df = pd.DataFrame(gbprice)
gbprice_df = gbprice_df.reset_index()
gbprice_df
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 

gbagg = df.groupby('neighbourhood_group')['price'].agg(['min', 'max'])
gbagg_df = pd.DataFrame(gbagg)
gbagg_df = gbagg_df.reset_index()
gbagg_df
```

```python
# 4. Using `groupby`, find the mean price for each room type in each neighbourhood_group.
gbroomtype = df.groupby('room_type')['price'].mean()
gbroomtype_df = pd.DataFrame(gbroomtype)
gbroomtype_df = gbroomtype_df.reset_index()
gbroomtype_df
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.
gbroomtype_all = df.groupby('room_type')['price'].agg(['min','max', 'mean', 'median','std'])
gbroomtype_all_df = pd.DataFrame(gbroomtype_all)
gbroomtype_all_df = gbroomtype_all_df.reset_index()
gbroomtype_all_df
```
